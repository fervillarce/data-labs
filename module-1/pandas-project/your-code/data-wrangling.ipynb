```python
import pandas as pd
import numpy as np
```

```python
# EXTRACT

# Importo los datos del archivo "GSAF5.csv", que está en el mismo directorio que este archivo.
# He hecho esto: data = pd.read_csv("GSAF5.csv"), pero me daba el error de UTF y he aplicado esta solución para poder leer el archivo.
df = pd.read_csv('GSAF5.csv', header=0, encoding = 'unicode_escape') # con 'latin-1' también lo abre.

# Echo un vistazo a una pequeña muestra de los datos.
df.head(30)
```

```python
# (3)
# Las columnas 'Case Number.1' y 'Case Number.2' parecen tener los mismos valores. Voy a comprobarlo.
count = 0
for ind in df.index:
    if df['Case Number.1'][ind] == df['Case Number.2'][ind]:
        count += 1
print(count)
```

```python
# Las columnas 'href formula' y 'href' parecen tener los mismos valores. Voy a comprobarlo.
count = 0
for ind in df.index:
    if df['href formula'][ind] == df['href'][ind]:
        count += 1
print(count)
# En el ejercicio de pipeline haré una función para hacer comparación de valores entre columnas.
```

```python
# Las columnas 'Case Number.1' y 'Case Number.2' tienen 5981 (de 5992 filas) valores iguales.
# Las columnas 'href formula' y 'href' tienen 5938 (de 5992 filas) valores iguales.
# Podemos concluir que son prácticamente redundantes.
# Voy a eliminar 'href' (porque tiene dos nulls más que href formula) y 'Case Number.2'
df.drop(['href','Case Number.2'], axis=1, inplace = True)
```

```python
# Voy a ver los diferentes valores de Sex.
df['Sex'].unique()
```

```python
# (4)
# Vistos los valores de Sex:
# 1 -> Las 'M ' las tengo que pasar a 'M'
# 2 -> 'lli', '.' y 'N', tengo que ver los nombres para saber si son male o female

# 1 -> Las 'M ' las tengo que pasar a 'M' con un replace. Hago un strip para eliminar espacios antes y delante de las strings.
df['Sex'] = df['Sex'].apply(lambda s: str(s).strip()) # Esta lambda podría aplicarla a todo el df directamente para que me lo aplique a todas las columnas.
```

```python
# 2 -> 'lli', '.' y 'N', tengo que ver los nombres para saber si son male o female
df[(df['Sex'] == 'lli') | (df['Sex'] == '.') | (df['Sex'] == 'N')].head()
```

```python
# El name de "lli" es de un male. Cambiamos "lli" por "M".
df['Sex'].replace('lli', 'M', inplace = True)
# El "." no se sabe qué es. 'N' es una pareja de hombre y mujer. Cambiamos "." y "N" por NaN.
df['Sex'].replace(['.','N'], np.NaN, inplace = True)
```

```python
# Quiero ver los tipos de cada uno de los campos, mediante dtypes o info()
df.dtypes
```

```python
# (5)
# Voy a ver los distintos valores de "Fatal (Y/N)" para proceder a limpiarlos si hay algún valor distinto a Y o N.
df['Fatal (Y/N)'].unique()
```

```python
df[(df['Fatal (Y/N)'] == 'UNKNOWN') | (df['Fatal (Y/N)'] == "#VALUE!") | (df['Fatal (Y/N)'] == 'n')]
```

```python
# El name de "lli" es de un male. Cambiamos "lli" por "M".
df['Fatal (Y/N)'].replace('UNKNOWN', np.NaN, inplace = True)
# El "." no se sabe qué es. 'N' es una pareja de hombre y mujer. Cambiamos "." y "N" por NaN.
df['Fatal (Y/N)'].replace('#VALUE!', np.NaN, inplace = True)
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].apply(lambda s: str(s).strip())
df['Fatal (Y/N)'].replace('F', 'Y', inplace = True)
```

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].apply(lambda s: str(s).upper())
```
