We would like to create a deciding factor column for each student. We standardize several columns and then pick the most important factor from a lookup table. If the standardized value is above 0.8, the student will be accepted.

We will start by creating a standardized column for `CGPA`, `GRE Score` and `LOR`. We will name these columns `CGPA_std`, `GRE_std`, and `LOR_std` respecively.

Recall that standardizing a column is done by subtracting the mean of the column from all observations in the column and then dividing each observation in the column by the column's standard deviation.

In the cell below, write a standardization function (a function that takes a column as input and returns a standardized column as output). Make sure to use the `numpy` versions of mean and standard deviation.

```python
def standardize(col):
    # This function takes a column from a dataframe and returns a standardized column by subtracting the column's mean
    # and dividing by the column's standard deviation.
    
    # Your code here:
    col_std = (col - np.mean(col))/np.std(col)
    return col_std
```

Now create the standardized columns `CGPA_std`, `GRE_std`, and `LOR_std` and add them to the `admissions` dataframe.

```python
# Your code here:
admissions['CGPA_std'] = standardize(admissions['CGPA'])
admissions['GRE_std'] = standardize(admissions['GRE Score'])
admissions['LOR_std'] = standardize(admissions['LOR'])
admissions.head()
```

We will generate the decision choice at random using the code below. Please run the cell.

```python
# Run this code:

from random import choices
std_columns = ['CGPA_std', 'GRE_std', 'LOR_std']

decision_choice = choices(std_columns, k=admissions.shape[0])
```

Now create the deciding column using the `lookup` function. The lookup column is `decision_choice` found above. Call the column resulting from the lookup function `deciding_column` and add it to the `admissions` dataframe.

```python
# Your code here:
admissions['deciding_column'] = admissions.lookup(admissions.index, decision_choice)
admissions.head()
```

Create a column called `decision` in the `admissions` dataframe. Assign 1 to this column if the value of `deciding_column` is greater than 0.8 and 0 otherwise.

```python
# Your code here:
admissions['decision'] = np.where(admissions['deciding_column'] > 0.8, 1, 0)
admissions.head()
```

How many applicants will be accepted to the program using the decision column? Compute the result below.

```python
# Your code here:
# admissions[admissions['decision'] == 1]['decision'].count()
print(admissions[admissions['decision'] == 1]['decision'].count(), "applicants will be accepeted.")
```

# Challenge 3 - Method Chaining

To increase our coding efficiency, let's make a number of changes to our dataframe in one line of code.

In the cell below, remove all non character symbols from the column names, replace all spaces in column names to underscores, and change all upper case characters to lower case. Assign these new column values to `admissions.columns`.

```python
# Your code here:
col_names = admissions.columns.copy()

import re # Importo re para poder sustituir los non-character symbols mediante regex.

new_col_names = []
for name in col_names:
    name = re.sub(r"\W", " ", name).replace(" ", "_").lower()
    new_col_names.append(name)
admissions.columns = new_col_names
admissions.head()

# También podría haber hecho 3 funciones distintas y aplicarlas en una líena a la lista de nombres de las columnas.
```

In the cell below, give all student with a university rating of 4 or higher a 10 point boost on their GRE score and split the column into 4 bins using the `cut` function. Assign this new score to the variable `adjusted_gre`.

```python
# Your code here:

# Give all student with a university rating of 4 or higher a 10 point boost on their GRE score
admissions['gre_score'] = np.where(admissions['university_rating'] >= 4, admissions['gre_score'] + 10 , admissions['gre_score'])
admissions.head()
```

```python
# Split the column into 4 bins using the cut function

# Como hay que dividir los valores en 4 bins, voy a utilizar los valores min, primer percentil, segundo percentil,
# tercer percentil, y max. Estos valores son los siguientes:

admissions['gre_score'].describe()
```

```python
admissions['adjusted_gre'] = pd.cut(admissions['gre_score'],
                                    [int(admissions['gre_score'].min()), int(admissions['gre_score'].quantile(0.25)),
                                    int(admissions['gre_score'].quantile(0.50)), int(admissions['gre_score'].quantile(0.75)), 
                                    int(admissions['gre_score'].max())])

admissions.head()
```

```python
# En lugar de dejar los intervalos numéricos, podríamos poner etiquetas:

admissions['adjusted_gre'] = pd.cut(admissions['gre_score'],
                                    [int(admissions['gre_score'].min()), int(admissions['gre_score'].quantile(0.25)),
                                    int(admissions['gre_score'].quantile(0.50)), int(admissions['gre_score'].quantile(0.75)), 
                                    int(admissions['gre_score'].max())],
                                    labels = ["low", "medium-low", "medium-high", "high"])

admissions.head()
```
